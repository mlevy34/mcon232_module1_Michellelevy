## Question 1 — Kebab-case to Title (Est. 10 minutes)

Write a function:

```python
def kebab_to_title(s: str) -> str:
    ...
```

It should:
- take a string like `"a-little-princess"`
- return `"A Little Princess"`

Rules:
- The input may have extra spaces around it (strip them).
- Words are separated by `-`.
- Use string methods and loops (no fancy tricks).

Test:
- `"  anne-of-green-gables  "` → `"Anne Of Green Gables"`

````python
# Q1: Write your function here

def kebab_to_title(s: str) -> str:
    s = s.strip().title()
    parts = s.split("-")
    new_sen = " ".join(parts)
    return new_sen


print(kebab_to_title("a-little-princess"))
print(kebab_to_title("  anne-of-green-gables  "))
````

## Question 2 — Count a Substring (Non-overlapping) (Est. 10 minutes)

Write a function:

```python
def count_substring(text: str, sub: str) -> int:
    ...
```

Return the number of times `sub` occurs in `text`, **without overlaps**.

Examples:
- `count_substring("aaaa", "aa")` → 2  (matches at 0–1 and 2–3)
- `count_substring("bananana", "ana")` → 2  (matches "ana" twice without overlaps)

````python
# Q2: Write your function here

def count_substring(text: str, sub: str) -> int:
    count = text.count(sub)
    return count


print(count_substring("aaaa", "aa"))
print(count_substring("bananana", "ana"))
````

## Question 3 — Phrase Palindrome (Est. 12–15 minutes)

Write a function:

```python
def is_phrase_palindrome(s: str) -> bool:
    ...
```

Return `True` if `s` is a palindrome **when you ignore**:
- spaces
- punctuation
- capitalization

Examples:
- `"Never odd or even"` → True
- `"Was it a rat I saw?"` → True
- `"Touro College"` → False

Hint: Build a cleaned string that keeps letters only.

````python
# Q3: Write your function here

def is_phrase_palindrome(s: str) -> bool:
  lst = []
  s = s.lower().strip()
  s= s.replace("?", " ")
  parts = s.split()
  for part in parts:
    if part != " ":
      lst.append(part)
  string = "".join(lst)
  if string[::-1] == string:
    return True
  else:
    return False



print(is_phrase_palindrome("Never odd or even"))
print(is_phrase_palindrome("Was it a rat I saw?"))
print(is_phrase_palindrome("Touro College"))
````

````python
# PROVIDED: Run this cell first to write students.csv

def write_students_csv(filename: str = "students.csv") -> None:
    header = ["name", "score", "city"]
    rows = [
        ["Chana", 92, "New York"],
        ["Miriam", 85, "Brooklyn"],
        ["Leah", 88, "Queens"],
        ["Rivka", 95, "Lakewood"],
        ["Sara", 90, "Baltimore"],
    ]

    fout = open(filename, "w", encoding="utf-8", newline="")
    fout.write(",".join(header) + "\n")
    for row in rows:
        pieces = []
        for item in row:
            pieces.append(str(item))
        fout.write(",".join(pieces) + "\n")
    fout.close()

write_students_csv("students.csv")
print("Wrote students.csv")
````
