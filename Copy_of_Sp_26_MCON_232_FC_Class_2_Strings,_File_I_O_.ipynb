```python
# CSV Library Example

import csv

def write_sample_books_csv(filename: str = "books.csv") -> None:
    """
    Writes a small, classroom-friendly books dataset to a CSV file.

    Creates/overwrites `filename`.
    """
    header = ["title", "author", "year", "country", "genre"]
    rows = [
        ["Little Women", "Louisa May Alcott", 1868, "United States", "Classic"],
        ["Anne of Green Gables", "L. M. Montgomery", 1908, "Canada", "Classic"],
        ["Heidi", "Johanna Spyri", 1881, "Switzerland", "Classic"],
        ["A Little Princess", "Frances Hodgson Burnett", 1905, "England", "Classic"],
        ["The Secret Garden", "Frances Hodgson Burnett", 1911, "England", "Classic"],
        ["Pollyanna", "Eleanor H. Porter", 1913, "United States", "Classic"],
        ["Black Beauty", "Anna Sewell", 1877, "England", "Classic"],
        ["Rebecca of Sunnybrook Farm", "Kate Douglas Wiggin", 1903, "United States", "Classic"],
        ["Eight Cousins", "Louisa May Alcott", 1875, "United States", "Classic"],
        ["What Katy Did", "Susan Coolidge", 1872, "United States", "Classic"],
    ]

    # newline="" is important on Windows/Colab to avoid blank lines in CSV output
    with open(filename, "w", encoding="utf-8", newline="") as file:
        writer = csv.writer(file)
        writer.writerow(header)
        writer.writerows(rows)


# Example usage in Colab:
# 1) Write the file
write_sample_books_csv("books.csv")


```

```python
def read_books_csv(filename : str="books.csv"):
  books = []
  fin = open(filename, "r")
  lines = fin.readlines()
  fin.close()

  #skip the header
  for line in lines[1:]:
    line = line.strip()
    if line == "":
      continue
    parts = line.split(",")
    title = parts[0].strip()
    author = parts[1].strip()
    year = int(parts[2].strip())
    country = parts[3].strip()
    genre = parts[4].strip()

    books.append((title, author, year, country, genre))
  return books
read_books_csv()
```

## Leetcode example  316

You are given a string s made of lowercase letters.  

Your task is to remove duplicate letters so that:  

- Each letter appears exactly once  

- The resulting string is the smallest possible in alphabetical (lexicographical) order  

- The relative order of letters must come from the original string
(you can’t rearrange freely).
- - Given these letters, I am looking for the smallest possible string, using each letter once, that can be formed by deleting characters, not reordering them. I can throw away letters if I know they appear again later
- - Two rules

- - - When scanning the string left to right:

- - - - You may discard a letter
- - - - - ✔ only if that letter appears again later

- - - - -  You prefer smaller letters earlier
✔ because they make the string lexicographically smaller
- - - - - Note : lexigraphic comparison only compares the first letter and ignores all others


Everything else follows from those two rules.


### Example 1
Input:  `s = "bcabc"`
Output: `"abc"`


Why?

You must keep only one of each letter: {a, b, c}

Among all valid results, "abc" is lexicographically smallest

## Example 2
Input:  `s = "cbacdcbc"`
Output: `"acdb"`
This is where the problem becomes non-trivial:

Simply keeping first occurrences gives "cbad" ❌

The correct solution must sometimes remove earlier letters to allow a smaller ordering later



### Inefficient and incorrect

### Efficient and correct

```python
# code here
s = "cbacdcbc"
result = []
used = []
for ch in s:
  if ch in used:
    continue

  while result and ch < result[-1] and result[-1] in s[s.index(ch)+1:]:
    removed = result.pop()
    used.remove(removed)
  result.append(ch)
  used.append(ch)
print("".join(result))
```
